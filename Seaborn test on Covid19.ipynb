```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("covid_19_india.csv")
data
```

```python
sns.pairplot(data[["ConfirmedIndianNational","ConfirmedForeignNational","Cured","Deaths","Confirmed"]])
```

```python
cor=data[["ConfirmedIndianNational","ConfirmedForeignNational","Cured","Deaths","Confirmed"]].corr()
cor
```

```python
sns.heatmap(cor,annot=True,cmap="plasma",vmin=1,vmax=1)
```

```python
g=sns.FacetGrid(data,col="State/UnionTerritory",col_wrap=5,height=5)
g=g.map(plt.plot, "ConfirmedForeignNational","ConfirmedIndianNational",marker="*")
```

```python
g=sns.FacetGrid(data,col="State/UnionTerritory",col_wrap=5,height=5)
g=g.map(plt.plot, "Confirmed","Deaths",marker="*")
```
