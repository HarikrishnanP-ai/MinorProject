```python
#linear Regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
maha
```

```python
#convert the date column into toordinates
maha["date"]=maha["date"].map(dt.datetime.toordinal)
maha.head()
```

```python
#getting depending and independent values
x=maha[["date"]]
y=maha[["confirmed"]]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3)
```

```python
print(xtrain.size)
print(ytrain.size)
```

```python
xtrain
```

```python
xtrain.shape#i dont wanna the columns
```

```python
xtrain.reshape(-1,1)
print(xtrain.shape)
```

# Hence, we use numpy library to fit the shape of our series into a desired shape

```python
model=LinearRegression()
```

```python
model.fit(np.array(xtrain).reshape(-1,1),np.array(ytrain).reshape(-1,1))
```

```python
ypred=model.predict(np.array(xtest).reshape(-1,1))
print(ypred)
```

```python
print(np.array(ytest).reshape(-1,1))
ytested=np.array(ytest).reshape(-1,1)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(ytested,ypred)
```

```python
maha.tail()
```

```python
#USING Decision Tree algo
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
```

```python
dtr.fit(np.array(xtrain).reshape(-1,1),np.array(ytrain).reshape(-1,1))
```

```python
ypred2=dtr.predict(np.array(xtest).reshape(-1,1))
```

```python
ypred2[0:5]
```
