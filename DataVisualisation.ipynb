```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data=pd.read_csv("covid_19_india.csv")
data.head(5)
```

```python
data.describe()
```

```python
data.info
```

```python
data1=data[["Date","State/UnionTerritory","Cured","Deaths","Confirmed"]]
data1.head()
data1.info
```

# Now take individual cases from the last day of august date=31/08/20

```python
lastD=data1[data1["Date"]=="11/09/20"]
lastD.head()
```

```python
#obtain the top 5 worst cases
lastD=lastD.sort_values(by="Confirmed",ascending=False)
worst=lastD[0:5]
worst
```

```python
#plot a bargraph statewise
sb.set(rc={'figure.figsize':(10,5)})
sb.barplot(x="Date",y="Confirmed",data=worst,hue="State/UnionTerritory")
plt.show()
```

```python
#the least affected states
least=lastD[-5:]#leave the end position empty if u want the last row too.
least
```

```python
sb.set(rc={'figure.figsize':(10,5)})
sb.barplot(x="Date",y="Confirmed",data=least,hue="State/UnionTerritory")
plt.show()
```

# We observe that Maharashtra has the worst number of cases till august,while Mizoram has the least number of cases

```python
maha=data1[data1["State/UnionTerritory"]=="Maharashtra"]
maha.head(12)
```

```python
maha.info
```

```python
maha["Transconf"]=np.log(maha["Confirmed"])
maha.head()
```

```python
plt.title("Plot of cases confirmed in Maharashtra")
plt.xlabel("Date")
plt.ylabel("Number of cases in Lakhs")
plt.scatter(maha.Date,maha.Confirmed)
```

```python
mahaf=maha[0:176]
mahaf
```

# plotting the total no. of cases till the last august

```python
sb.set(rc={'figure.figsize':(25,15)})
sb.lineplot(x="Date",y="Confirmed",data=mahaf)
plt.show()
```

# Plotting the deaths v/s confirmed

```python
sb.set(rc={'figure.figsize':(15,10)})
sb.lineplot(x="Confirmed",y="Deaths",data=mahaf,color="r")
plt.show()
```

# plotting the cure v/s confirmed in Maharashtra

```python
sb.set(rc={'figure.figsize':(15,10)})
sb.lineplot(x="Cured",y="Confirmed",data=mahaf,color="g")
plt.show()
```
